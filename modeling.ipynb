```python
import numpy as np
import pandas as pd
```

```python
scaled_poly_cont = pd.read_csv('scaled_poly_cont.csv')
scaled_poly_all = pd.read_csv('scaled_poly_all.csv')
print(scaled_poly_cont.head())
print(scaled_poly_all.head())
```

```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold, train_test_split, cross_val_score
from sklearn.metrics import log_loss, f1_score, classification_report
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
```

```python
#modeler
def modeler (data, model):
    X_train, X_test, y_train, y_test = train_test_split(
        data.drop(['G3', 'grade_rank'], axis=1), data['grade_rank'],
        test_size=.2, random_state=2022
        )
    print(model)
    model.fit(X_train, y_train)

    pred = model.predict(X_test)

    print(f1_score(pred, y_test, average='weighted'))
    print(classification_report(pred, y_test))
```

```python
modeler(data=scaled_poly_cont, model= LogisticRegressionCV())
modeler(data=scaled_poly_all, model= LogisticRegressionCV())
```

```python
modeler(data=scaled_poly_cont, model= DecisionTreeClassifier())
modeler(data=scaled_poly_all, model= DecisionTreeClassifier())
```

## Hyperparameter Tunning

```python
X_train, X_test, y_train, y_test = train_test_split(
        scaled_poly_all.drop(['G3', 'grade_rank'], axis=1), scaled_poly_all['grade_rank'],
        test_size=.2, random_state=2022
        )
```

```python
#xgboot
import optuna
def objective(trial):
    classifier = 'catb'

    if classifier == 'catb':
        param = {}
        param['learning_rate'] = trial.suggest_discrete_uniform("learning_rate", 0.001, 0.01, 0.001)
        param['depth'] = trial.suggest_int('depth', 10, 20)
        param['l2_leaf_reg'] = trial.suggest_discrete_uniform('l2_leaf_reg', 1.0, 5.5, 0.5)
        param['min_child_samples'] = trial.suggest_categorical('min_child_samples', [1, 4, 8, 16, 32, 64])
        param['iterations'] = trial.suggest_discrete_uniform("iterations", 200, 2000, 50)
        param['use_best_model'] = True
        param["colsample_bylevel"] =  trial.suggest_float("colsample_bylevel", 0.001, 0.1)
        param['random_state'] = 2022
        param['scale_pos_weight'] = trial.suggest_int('scale_pos_weight', 1, 10)
        param['bootstrap_type'] = trial.suggest_categorical("bootstrap_type", ["Bayesian", "Bernoulli", "MVS"])
        param['logging_level'] = 'Silent'

        clf = XGBClassifier(**param)


    return cross_val_score(clf, X_train, y_train, n_jobs = -1, cv=5).mean()


```

```python
#creating a study
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50)
```

```python
trial = study.best_trial

print('Accuracy: {}'.format(trial.value))
print("Best hyperparameters: {}".format(trial.params))
```

```python
#random forest
trial = study.best_trial

print('Accuracy: {}'.format(trial.value))
print("Best hyperparameters: {}".format(trial.params))
```

```python
#LGBM
def objective(trial):
    classifier = 'lgB'

    if classifier == 'lgB':
        param = {}
        param["learning_rate"] = trial.suggest_float("learning_rate", 1e-2, 0.25, log=True)
        param['subsample'] = trial.suggest_float("subsample", 0.1, 1.0)
        param['colsample_bytree'] = trial.suggest_float("colsample_bytree", 0.1, 1.0)
        param['max_depth'] = trial.suggest_int("max_depth", 5, 20)
        param['scale_pos_weight'] = trial.suggest_int('scale_pos_weight', 3, 10)

        clf = LGBMClassifier(**param)


    return cross_val_score(clf, X_train, y_train, n_jobs = -1, cv=5).mean()




#creating a study
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50)
```

```python
trialcat = study.best_trial

print('Accuracy: {}'.format(trialcat.value))
print("Best hyperparameters: {}".format(trialcat.params))
```

```python
#curating the model with tuneded parameters
logreg = LogisticRegressionCV()
randf = RandomForestClassifier(n_estimators= 1800, criterion ='gini', min_samples_split=4,
                                min_samples_leaf= 2, min_weight_fraction_leaf= 0.0145002008293025, bootstrap=True)
catb = CatBoostClassifier()
xgbx = XGBClassifier(learning_rate=0.017, depth=11, l2_leaf_reg=2.0, min_child_samples=32,
                      iterations=1650.0, colsample_bylevel= 0.0997976838420612, scale_pos_weight=6,
                      bootstrap_type= 'Bernoulli')
lgb = LGBMClassifier(learning_rate = 0.019543456860390714, subsample = 0.9019992236460949,
                     colsample_bytree= 0.7096552668749082, max_depth=16, scale_pos_weight=5)
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import StackingClassifier
import matplotlib.pyplot as plt

# get a stacking ensemble of models
def get_stacking():
	# define the base models
	level0 = list()
	level0.append(('lr', LogisticRegressionCV()))
	level0.append(('randf', RandomForestClassifier()))
	level0.append(('cat', CatBoostClassifier()))
	level0.append(('xgb', XGBClassifier()))
	level0.append(('lgb', LGBMClassifier()))
	# define meta learner model
	level1 = LogisticRegressionCV()
	# define the stacking ensemble
	model = StackingClassifier(estimators=level0, final_estimator=level1, cv=5)
	return model

# get a list of models to evaluate
def get_models():
	models = dict()
	models['lr'] = LogisticRegressionCV()
	models['randf'] = RandomForestClassifier()
	models['cat'] = CatBoostClassifier()
	models['xgb'] = XGBClassifier()
	models['lgb'] = LGBMClassifier()
	models['stacking'] = get_stacking()
	return models

# evaluate a give model using cross-validation
def evaluate_model(model, X, y):
	cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
	scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
	return scores


# get the models to evaluate
models = get_models()
# evaluate the models and store results
results, names = list(), list()
for name, model in models.items():
	scores = evaluate_model(model, X_train, y_train)
	results.append(scores)
	names.append(name)
	print('>%s %.3f (%.3f)' % (name, np.mean(scores), np.std(scores)))
# plot model performance for comparison
plt.boxplot(results, labels=names, showmeans=True)
plt.show()
```
